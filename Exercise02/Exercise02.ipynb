```python
import numpy as np
```

```python
def sigmoid(x): 
    return 1 / (1 + np.exp(-x))
```

```python
sigmoid(-1)
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 0.1)
x
```

```python
plt.plot(x, sigmoid(x))
plt.show()
```

```python
def tanh(x): 
    return 2 / (1 + np.exp(-2*x)) - 1
```

```python
plt.plot(x, tanh(x))
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.plot(x, relu(x))
plt.show()
```

```python
def softmax(list): 
    return np.exp(list) / np.sum(np.exp(list))
```

```python
result = softmax( [0, 1, 168, 8, 2]) 
result
```
